```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Step 2: Draw samples"""
def draw_sample(alpha, beta):
    sample = np.random.beta(alpha,beta)
    print("Samples:",sample)
    return sample

"""Step 3: Choose highest sample"""
def choose_action(samples):
    action = np.argmax(samples)
    print("Action chosen:", action)
    return action

"""Step 4: Get reward"""
def get_reward(action):
    if np.random.uniform() < theta[action]:
        print("Reward: 1")
        return 1
    
    else:
        print("Reward: 0")
        return 0

"""Step 5: Calculate posterior"""
def update_posterior(reward):
    alpha[action] += reward
    beta[action] += 1 - reward
    print("Updated alpha and beta:", alpha, beta)

def plot_distribution(alpha,beta):
    x = np.linspace(0,1, 1000)
    for i in range(3):
        plt.plot(x, scipy.stats.beta.pdf(x,alpha[i],beta[i]), label=f"Arm {i}")
    
    plt.grid()
    plt.legend()
    plt.show()
```

```python
"""Step 1: Choose prior"""
theta = [0.5, 0.6, 0.4]
alpha = [1, 1, 1]
beta = [2, 1, 1]
```

```python
"""Step 2: Draw samples"""
samples = draw_sample(alpha,beta)
```

```python
"""Step 3: Choose highest sample"""
action = choose_action(samples)
```

```python
"""Step 4: Get reward"""
reward = get_reward(action)
```

```python
"""Step 5: Calculate posterior"""
update_posterior(reward)
```

```python
"""Plot distributions"""
plot_distribution(alpha, beta)
```

```python
for i in range(100):
    samples = draw_sample(alpha,beta)
    action = choose_action(samples)
    reward = get_reward(action)
    update_posterior(reward)
    
plot_distribution(alpha, beta)
```
